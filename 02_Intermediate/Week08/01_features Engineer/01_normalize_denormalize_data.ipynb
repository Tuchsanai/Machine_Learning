













# Demo MinMaxScale

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
data   = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]




xdata = pd.DataFrame(data, columns=['x1', 'x2'])
xdata
```

# MinMaxScaler

```python
scaler = MinMaxScaler()
xscale = scaler.fit_transform(xdata)
```

```python
xscale = pd.DataFrame(xscale, columns=['x1', 'x2'])
xscale
```

# inverse MinMaxScaler to xdata

```python
x = scaler.inverse_transform(xscale)
x = pd.DataFrame(x,  columns=['x1', 'x2'])
x
```

# Demo of StandardScale

```python
from sklearn.preprocessing import StandardScaler
    
import pandas as pd
data   = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = StandardScaler()


xdata = pd.DataFrame(data, columns=['x1', 'x2'])
xdata
```

```python
xscale = scaler.fit_transform(xdata)
xscale = pd.DataFrame(xscale, columns=['x1', 'x2'])
xscale
```

```python
x = scaler.inverse_transform(xscale)
x = pd.DataFrame(x,  columns=['x1', 'x2'])
x
```

```python

# Import the necessary libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
    
```

```python

# Load the dataset
url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv'
df = pd.read_csv(url)


df.info()
```

```python

# Preview the data
df.head()
```

# Selected all numerical Column

```python
df_nums = df.select_dtypes(exclude='object')

df_nums
```

```python
df_nums.columns.tolist()
```

# Min-Max scaling


```python

# Columns to normalize
columns_to_normalize = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']
    
```

```python

# Min-Max scaling
min_max_scaler = MinMaxScaler()
df_min_max = df.copy()
df_min_max[columns_to_normalize] = min_max_scaler.fit_transform(df[columns_to_normalize])

# Preview the data after Min-Max scaling
df_min_max.head()

    
```

```python

# De-normalize the data
df_min_max_denormalized = df_min_max.copy()

df_min_max_denormalized[columns_to_normalize] = min_max_scaler.inverse_transform(df_min_max[columns_to_normalize])

# Preview the data after de-normalization
df_min_max_denormalized.head()
```

# Z-score normalization


```python

# Columns to normalize
columns_to_normalize = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']
    
# Standard scaling (Z-score normalization)
standard_scaler = StandardScaler()
df_standard = df.copy()
df_standard[columns_to_normalize] = standard_scaler.fit_transform(df[columns_to_normalize])

# Preview the data after Standard scaling
df_standard.head()
```

```python

# De-normalize the data
df_standard_denormalized = df_standard.copy()
df_standard_denormalized[columns_to_normalize] = standard_scaler.inverse_transform(df_standard[columns_to_normalize])

# Preview the data after de-normalization
df_standard_denormalized.head()
    
```
